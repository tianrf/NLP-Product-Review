```python
def scrape_taobao(url_p1,url_p2,header,comment_name,csv_name,max_page):
    with open(csv_name,'w',encoding='utf-8-sig') as csv_file:
        csv_writer=writer(csv_file)
        headers=['comment_sen','score']
        csv_writer.writerow(headers)
        for i in range(max_page):
            url=url_p1+str(i)+url_p2
            try:
                html = requests.get(url, headers=header).text
                json_text = re.search('^[^(]*?\((.*)\)[^)]*$', html).group(1)
                text = json.loads(json_text)
            except:
                continue
            for comment in text['comments']:
                try:
                    comment_sen=comment['content']
                    score=comment['score']
                    csv_writer.writerow([comment_sen,score])
                except:
                    print('some information lost!')
                

    
    
```
